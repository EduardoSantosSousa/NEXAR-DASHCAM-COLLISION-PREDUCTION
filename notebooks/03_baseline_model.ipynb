# 03 - Baseline Model

This notebook documents the first frame-level baseline for the Nexar Collision Prediction project.

## Baseline Question

Can individual sampled frames provide enough visual signal to distinguish collision or near-collision risk from normal driving?

````python
from pathlib import Path
import json

import pandas as pd
from IPython.display import Image, display

PROJECT_ROOT = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
metrics_path = PROJECT_ROOT / "models" / "reports" / "baseline_metrics.json"
frame_predictions_path = PROJECT_ROOT / "outputs" / "predictions" / "baseline_frame_predictions.csv"
video_predictions_path = PROJECT_ROOT / "outputs" / "predictions" / "baseline_video_predictions.csv"
figures_dir = PROJECT_ROOT / "outputs" / "figures"

metrics_path
````

## Training Command

The current baseline was trained with:

```powershell
.\venv\Scripts\python.exe scripts\train_baseline.py --epochs 3 --batch-size 16
```

````python
metrics = json.loads(metrics_path.read_text(encoding="utf-8"))

metrics["device"], metrics["cuda_available"], metrics["train_videos"], metrics["val_videos"]
````

````python
pd.DataFrame(metrics["history"])
````

````python
summary = pd.DataFrame(
    [
        {"level": "frame", **metrics["frame_metrics"]},
        {"level": "video_mean", **metrics["video_metrics_mean"]},
        {"level": "video_max", **metrics["video_metrics_max"]},
    ]
)
summary
````

````python
frame_predictions = pd.read_csv(frame_predictions_path, dtype={"id": str})
video_predictions = pd.read_csv(video_predictions_path, dtype={"id": str})

display(frame_predictions.head())
display(video_predictions.head())
````
