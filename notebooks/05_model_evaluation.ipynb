# 05 - Temporal Alert Evaluation

This notebook analyzes temporal risk scores, predicted alert timestamps, and alert-specific KPIs.

## Core Question

Can the model emit an alert before the event occurs, and how does the chosen risk threshold affect recall, false alarms, and alert timing?

```python
from pathlib import Path
import json

import pandas as pd
from IPython.display import Image, display

PROJECT_ROOT = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
alert_metrics_path = PROJECT_ROOT / "models" / "reports" / "alert_metrics.json"
threshold_sweep_path = PROJECT_ROOT / "models" / "reports" / "alert_threshold_sweep.csv"
risk_scores_path = PROJECT_ROOT / "outputs" / "predictions" / "temporal_risk_scores.csv"
alert_predictions_path = PROJECT_ROOT / "outputs" / "predictions" / "alert_predictions.csv"
figures_dir = PROJECT_ROOT / "outputs" / "figures"
```

## Temporal Alert Metrics at Threshold 0.5

```python
alert_report = json.loads(alert_metrics_path.read_text(encoding="utf-8"))
alert_report["device"], alert_report["metrics"]
```

```python
risk_scores = pd.read_csv(risk_scores_path, dtype={"id": str})
alert_predictions = pd.read_csv(alert_predictions_path, dtype={"id": str})

display(risk_scores.head())
display(alert_predictions.head())
```

## Threshold Sweep

A fixed threshold of 0.5 is only an initial operating point. The sweep below helps identify the trade-off between recall and false alarms.

```python
threshold_sweep = pd.read_csv(threshold_sweep_path)
threshold_sweep[["threshold", "alert_precision", "alert_recall", "false_alarm_rate", "missed_event_rate", "mean_predicted_lead_time", "mean_alert_time_error"]]
```

## Interpretation Notes

The current model should be treated as an early temporal baseline. If alert recall is high but false alarm rate is also high, the next step is to improve calibration, use transfer learning, and train with more temporally meaningful labels.
